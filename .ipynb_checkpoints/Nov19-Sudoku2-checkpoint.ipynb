## New Solution
You can create a new solution from a known solution by
* __Mutate__: Swap one number for another.  Ex.  Turn the 7's into 3's and the 3's into 7's.
* __Flip__: Flip horizontally or vertically.
* __Rotate__: Rotating puzzle 90, 180, 270 degrees
* __Swap__: Swapping rows (as long as they are in the same blocks). Swapping columns (as long as they are in the same blocks)
* __Flop__: Swap all three rows in a block with another (or columns)

To create a new puzzle: Flip, Flop, Swap, Rotate and Mutate.

```python
n1 = 3; n2 = 7 # numbers to be swapped
dct = PuzzleSolved()
sudoku.PrintSolution(dct)
print('')
groups = sudoku.Architecture1()
for i in range( 81 ):
    if dct[i][0] == n1:
        dct[i][0] =-2
for i in range( 81 ):
    if dct[i][0] == n2:
        dct[i][0] = n1
for i in range( 81 ):
    if dct[i][0] == -2:
        dct[i][0] = n2
sudoku.PrintSolution(dct)
```

```python
# FLIP - only for Architecture 1
mat = np.zeros((9,9),int)
# put numbers in matrix
k = 0
for i in range(9):
    for j in range( 9 ):
        mat[i,j] = dct[k][0]
        k += 1
# flip it
mat = np.fliplr(mat)
# put back into puzzle
k = 0
for i in range(9):
    for j in range(9):
        dct[k][0] = mat[i,j]
        k += 1
sudoku.PrintSolution(dct)
```

```python
# Rotate 
mat = np.zeros((9,9),int)
# put numbers in matrix
k = 0
for i in range(9):
    for j in range( 9 ):
        mat[i,j] = dct[k][0]
        k += 1
# flip it
mat = np.rot90(mat) # 180, 270
# put back into puzzle
k = 0
for i in range(9):
    for j in range(9):
        dct[k][0] = mat[i,j]
        k += 1
sudoku.PrintSolution(dct)
```

```python
# swap rows (within same blocks)
# FLIP - only for Architecture 1
mat = np.zeros((9,9),int)
# put numbers in matrix
k = 0
for i in range(9):
    for j in range( 9 ):
        mat[i,j] = dct[k][0]
        k += 1
#
row1 = 0; row2 = 2
temp = mat[row1]+0
mat[row1] = mat[row2]+0
mat[row2] = temp +0
# put back into puzzle
k = 0
for i in range(9):
    for j in range(9):
        dct[k][0] = mat[i,j]
        k += 1
sudoku.PrintSolution(dct)
```

```python
# swap rows (within same blocks)
# FLOP - only for Architecture 1
mat = np.zeros((9,9),int)
# put numbers in matrix
k = 0
for i in range(9):
    for j in range( 9 ):
        mat[i,j] = dct[k][0]
        k += 1
#
brow1 = 0; brow2 = 3
temp = mat[brow1:brow1+3]+0
mat[brow1:brow1+3] = mat[brow2:brow2+3]+0
mat[brow2:brow2+3] = temp +0
# put back into puzzle
k = 0
for i in range(9):
    for j in range(9):
        dct[k][0] = mat[i,j]
        k += 1
sudoku.PrintSolution(dct)
```

## The First Puzzle
The previous works well if you already have a solution, but what about the first puzzle?

Look for a pattern

$$\begin{matrix} 
1 & 2 & 3 && 4 & 5 & 6 && 7 & 8 & 9 \\
4 & 5 & 6 && 7 & 8 & 9 && 1 & 2 & 3 \\
7 & 8 & 9 && 1 & 2 & 3 && 4 & 5 & 6 \\
& & & & & && & && \\
2 & 3 & 4 && 5 & 6 & 7 && 8 & 9 & 1 \\
5 & 6 & 7 && 8 & 9 & 1 && 2 & 3 & 4 \\
8 & 9 & 1 && 2 & 3 & 4 && 5 & 6 & 7 \\
& & & & & && & && \\
3 & 4 & 5 && 6 & 7 & 8 && 9 & 1 & 2 \\
6 & 7 & 8 && 9 & 1 & 2 && 3 & 4 & 5 \\
9 & 1 & 2 && 3 & 4 & 5 && 6 & 7 & 8 \\
\end{matrix}$$

The _theory_ is that through the right combinations of _Flip, Flop, Swap, Rotate and Mutate_, it is possible to convert this solution to any other puzzle.
